# Problem 02

```python
student = pd.read_csv('students.csv')

student
```

```python
sns.scatterplot(data=student, x='hours_studied', y='marks', hue='gender')
```

# Problem 03

```python
sns.relplot(data = student, x='hours_studied', y='marks', col='subject')
```

# Problem 04

```python
sns.displot(kind = 'hist', data = student, x='marks', kde=True)
```

# Problem 05

```python
sns.kdeplot(data=student, x='hours_studied')
```

# Problem 06

```python
sns.countplot(data = student, x = 'subject', hue='gender')
```

# Problem 07

```python
sns.regplot(data = student, x = 'hours_studied', y = 'marks')
```

```python
sns.pairplot(data = student)
```
