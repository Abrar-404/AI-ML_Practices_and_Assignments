# Problem 01

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Practice.csv')

df
```

```python
df['Discounted_Price'] = df['Price'] * 0.90

df.to_csv('discounted_products.csv', index=False)

df
```

# Problem 02

```python
print('Total Duplicates:', df['EmployeeID'].duplicated().sum())

df_cleaned = df.drop_duplicates(subset='EmployeeID')

df_cleaned
```

# Problem 03

```python
print('TOTAL MISSING VALUES:\n', df[['Age', 'Salary']].isna().sum())

avg_age = df['Age'].mean()
df['Age'].fillna(avg_age, inplace=True)

df.dropna(subset='Salary', inplace=True)

df
```

# Problem 04

```python
print('Avg Marks:\n', df[['Marks1', 'Marks2', 'Marks3']].mean())
print()
print ('Max Marks:\n', df[['Marks1', 'Marks2', 'Marks3']].max())


df['Total Marks'] = df[['Marks1', 'Marks2', 'Marks3']].sum(axis=1)
print()
print('Standard Deviation:\n', df['Total Marks'].std())

print()
df
```

# Problem 05

```python
# Histogram based on age

plt.hist(df['Age'])
plt.xlabel('Age')
plt.ylabel('Frequency of Employee')
plt.title('Age Chart of Employees')

plt.show()
```

```python
# Salary and Age

plt.scatter(df['Age'], df['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs Salary')

plt.grid()
plt.show()
```

```python
# Price and Discounted Price

plt.scatter(df['Price'], df['Discounted_Price'])
plt.xlabel('Price')
plt.ylabel('Discounted Price')
plt.title('Price vs Discounted Price')

plt.grid()
plt.show()
```

# Problem 06

```python
iq_df = pd.read_csv('student_IQdata.csv')

iq_df
```

```python
#Study Hour vs IQ Score

plt.scatter(iq_df['Study_Hour'], iq_df['IQ_Score'])
plt.xlabel('Study Hour')
plt.ylabel('IQ_Score')
plt.title('Study Hour vs IQ Score')
plt.grid()
plt.show()

# Ans: Positively Correlated
```

```python
# Shoe Size vs Study Hour

plt.scatter(iq_df['Shoe_Size'], iq_df['Study_Hour'])
plt.xlabel('Shoe Size')
plt.ylabel('Study Hour')
plt.title('Shoe Size vs Study Hour')
plt.grid()
plt.show()

# Ans: Not Correlated
```

```python
# Histogram of shoe size

plt.hist(iq_df['Shoe_Size'], edgecolor='black')
plt.xlabel('Shoe Size')
plt.ylabel('Frequency of Students')
plt.title('Histogram of Shoe Size of Students')

plt.show()
```
