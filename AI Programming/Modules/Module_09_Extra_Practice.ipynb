# Problem 03

```python
arr = np.arange(10, 50, 2)
print(arr)

# Using linspace
arr_line = np.linspace(10, 48, 20)
print(arr_line)

# Using range with list comprehension
arr_loop = np.array([x for x in range(10, 50, 2)])
print(arr_loop)
```
