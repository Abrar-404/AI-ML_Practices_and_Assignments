```python
import numpy as np
```

```python
# Dataset

data = np.array([
    [9.5, 0.5, 95.5],
    [8.2, 1.2, 86.4],
    [7.5, 1.5, 82.1],
    [7.0, 1.0, 81.0],
    [6.5, 2.0, 74.5],
    [5.0, 1.5, 70.0],
    [4.5, 3.0, 60.5],
    [3.0, 2.5, 55.0],
    [2.0, 4.0, 44.0],
    [1.5, 4.5, 39.5]
])

data


# Randomizing the dataset

random_index = np.random.choice(10, size=(10,), replace = False)

data = data[random_index]

data


# feature, target extraction

X = data[ : , :2]

y = data[ : , 2:]

y


# train, test split

train_index = int(len(data) * 0.8)

# train feature
X_train = X[ : train_index, : ]
y_train = y[ : train_index, : ]

# test feature
X_test = X[train_index : , ]
y_test = y[train_index : ,]

X_test
```
