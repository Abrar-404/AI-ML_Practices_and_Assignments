```python
import numpy as np
import pandas as pd

np.random.seed(24)

num_employees = 50  # you can increase this if needed

# Employee data
emp_ids = np.arange(101, 101 + num_employees)
emp_ages = np.random.randint(20, 55, num_employees)
emp_income = np.random.randint(35000, 95000, num_employees)

# Fixed set of roles (repetition allowed)
roles_list = [
    "HR Manager",
    "Software Developer",
    "Accountant",
    "Marketing Executive",
    "Sales Executive",
    "Backend Developer",
    "HR Executive",
    "Financial Analyst",
    "Digital Marketer",
    "Operations Manager"
]

emp_roles = np.random.choice(roles_list, num_employees)

# Create DataFrame
data = pd.DataFrame({
    "emp_id": emp_ids,
    "emp_age": emp_ages,
    "emp_salary": emp_income,
    "role": emp_roles
})

# Save dataset
data.to_csv("staff_data.csv", index=False)
```

# Q1

```python
# Employee Age
emp_age_array = np.array(data["emp_age"])
print(f"Employee Age Array: {emp_age_array}")

print("Data Type:", emp_age_array.dtype)
print("Number of Dimensions:",emp_age_array.ndim)
print("Shape:",emp_age_array.shape)
print("Total Number of Elements:",emp_age_array.size)


# Employee Salary
emp_salary_array = np.array(data["emp_salary"])
print(f"Employee Salary Array: {emp_salary_array}")

print("Data Type:", emp_salary_array.dtype)
print("Number of Dimensions:",emp_salary_array.ndim)
print("Shape:",emp_salary_array.shape)
print("Total Number of Elements:",emp_salary_array.size)
```

# Q2

```python
salary_array = np.array(data['emp_salary'])
age_array = np.array(data['emp_age'])

max_salary = salary_array.max()
print(f"Maximum Salary: {max_salary}")

min_salary = salary_array.min()
print(f"Minimum Salary: {min_salary}")

avg_salary = salary_array.mean()
print(f'Average Salary: {avg_salary}')

avg_age = age_array.mean()
print(f'Average Age: {avg_age}')
```

# Q3

```python
emp_age_array = np.array(data["emp_age"])
emp_salary_array = np.array(data["emp_salary"])

result = (emp_age_array >= 25) & (emp_age_array <= 40) & (emp_salary_array > 50000)

filtered_age = emp_age_array[result]
filtered_salary = emp_salary_array[result]

emp_count = np.count_nonzero(result)

print(f"Filtered Age: {filtered_age}")
print(f"Filtered Salary: {filtered_salary}")
print(f'Total Count: {emp_count}')
```

```python
emp_age_array = np.array(data["emp_age"])
print(f"Original Array: {emp_age_array}")

average_age = emp_age_array.mean()
print(f"Average Age: {average_age}")

updated_array = np.where((emp_age_array < average_age), (emp_age_array + 4), (emp_age_array + 2))
print(f"Updated Array: {updated_array}")
```

# Q5

```python
employee_roles = np.array(data['role'])
employee_salary = np.array(data['emp_salary'])

roles = ["Software Developer", "Backend Developer", "Digital Marketer"]
integrate_roles = np.isin(employee_roles, roles)

old_total_salary = employee_salary.sum()
print(f"Old total salary: {old_total_salary}")

copied_salary = employee_salary.copy()
increased = employee_salary[integrate_roles] * 0.08
copied_salary[integrate_roles] = employee_salary[integrate_roles] + increased
new_salary = copied_salary

new_total_salary = new_salary.sum()
print(f'New Total Salary: {new_total_salary}')

differences = new_total_salary - old_total_salary
print(f'Salary Differences: {differences}')
```

```python
previous_salary = data['emp_salary']

rate1 = previous_salary * 0.12
rate2 = previous_salary * 0.05

increased_dev_sal = previous_salary + rate1
increased_other_sal = previous_salary + rate2

data['previous_salary'] = previous_salary
data['new_salary'] = increased_other_sal

data.loc[data['role'].isin(['Software Developer', 'Backend Developer']), 'new_salary'] = increased_dev_sal

new_salary = data['new_salary']

data['increase_of_salary'] = new_salary - previous_salary

output = data[['role', 'previous_salary' , 'new_salary' , 'increase_of_salary']]
print(output)
```
