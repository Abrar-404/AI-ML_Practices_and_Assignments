# Task 2

```python
# total number of missing values per column
print(df.isnull().sum())

print()

# Which feature contains the highest number of missing values?
print('Feature with most missing values:', df.isnull().sum().idxmax())
print('Number of missing values:', df.isnull().sum().max())

# ans:- sex with 10 missing values
```

# Task 3

```python
# Check if there are any duplicate rows in the penguins dataset. How many duplicates are there?

print('Duplicate Rows:- \n', df.duplicated())
print('Total Duplicates:', df.duplicated().sum())

# Write the code to permanently remove any duplicate rows from your DataFrame.
df.drop_duplicates(inplace=True)
```

# Task 4

```python
# Use the appropriate method to print a concise summary of the DataFrame, including the data types of each column and non-null counts.
df.info()

print()
print()

# Generate a statistical summary for the numerical columns.
print('Statistical Summary: \n',df.describe())

print()
print()

# Based on your statistical summary, what is the maximum body_mass_g of a penguin in this dataset? What is the average (mean) flipper_length_mm?
print('Maximum Body Mass:-', round(df['body_mass_g'].max(), 2))
print('Average Flipper Length:-', round(df['flipper_length_mm'].mean(), 2))
```

# Task 5

```python
# Use Seaborn to create a count plot for the species column.
sns.countplot(data=df, x='species')
plt.title('Distribution of Penguin Species')   # good practice to add title
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()

print()
print()

# Which species is the most common in this dataset?
print('Total Species:- \n', df['species'].value_counts())

print()
print()

print('Most Common Species:-', df['species'].value_counts().idxmax())
# ans:- Adelie is the most common species with 152 occurrences.
```

# Task 6

```python
# Write the code to calculate the percentage distribution of the island column

round((df['island'].value_counts() / len(df)) * 100, 2)
```

# Task 7

```python
# Extract the value counts for the sex column.
print(df['sex'].value_counts())

# Create a pie chart using Matplotlib showing the male vs. female distribution.
# Ensure the pie chart includes percentage labels with one decimal point (example 50.5%).
# Add a title to your chart.
sex_counts = df['sex'].value_counts()

# Clean: filter out invalid entries before plotting
sex_counts = sex_counts[sex_counts.index.isin(['Male', 'Female'])]  # keep only valid
sex_labels = sex_counts.index

plt.title('Male vs Female Penguin Distribution')

plt.pie(sex_counts, labels = sex_labels, autopct = '%1.1f%%')
plt.show()
```
