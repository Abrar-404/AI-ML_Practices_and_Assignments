# 0) Starter Code (Run First)

```python
# You must run this code and use this dataset for all questions
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
num_students = 50

# Student data
student_ids = np.arange(201, 201 + num_students)
student_ages = np.random.randint(15, 22, num_students)
student_scores = np.random.randint(40, 100, num_students)
student_attendance = np.random.randint(60, 100, num_students)  # percentage

departments_list = [
    "Science",
    "Commerce",
    "Arts",
    "Computer Science",
    "Mathematics",
    "English",
    "Physics",
    "Chemistry",
    "Biology",
    "Economics"
]

student_departments = np.random.choice(departments_list, num_students)

grade_conditions = [
    student_scores >= 80,
    (student_scores >= 65) & (student_scores < 80),
    (student_scores >= 50) & (student_scores < 65),
    student_scores < 50
]
grade_values = ["A", "B", "C", "F"]
student_grades = np.select(grade_conditions, grade_values, default="F")

# Create DataFrame
data = pd.DataFrame({
    "student_id": student_ids,
    "student_age": student_ages,
    "score": student_scores,
    "attendance": student_attendance,
    "department": student_departments,
    "grade": student_grades
})

# Save dataset
data.to_csv("school_data.csv", index=False)
```

# Question 01: [ Marks 10 ]

Use NumPy arrays from the score and attendance columns. Print both arrays and display:

* Data type
* Number of dimensions
* Shape
* Total number of elements

# Write Answer 01:


```python
score_array = np.array(data['score'])
attendance_array = np.array(data['attendance'])

print('Score Array:', score_array, '\n')
print('Attendance Array:', attendance_array, '\n')

print('Score Array Data Type:', score_array.dtype)
print('Attendace Array Data Type:', attendance_array.dtype, '\n')

print('Score Array Number of dimensions:', score_array.ndim)
print('Attendance Array Number of dimensions:', attendance_array.ndim, '\n')

print('Score Array Shape:', score_array.shape)
print('Attendance Array Shape:', attendance_array.shape, '\n')

print('Score Array Total Number of elements:', score_array.size)
print('Attendance Array Total Number of elements:', attendance_array.size)
```

# Question 02: [ Marks 10 ]

* Find the maximum and minimum score
* Calculate the average score
* Calculate the average attendance

Then, extract students who:

Print:


# Question 04: [ Marks 10 ]

##
Using .loc:

* Increase the score by 10% for students in "Computer Science" and "Mathematics" departments
* Increase the score by 5% for all other departments
* Cap the score at 100 (no score should exceed 100 after the raise)

Then print a DataFrame that includes:

* student_id, department, previous_score, new_score, score_increase


## Write Answer 04:

```python
df_copy = df.copy()
df_copy['previous_score'] = df_copy['score']

cs_and_math_dept = df_copy['department'].isin(['Computer Science', 'Mathematics'])

increase_cs_and_math = df_copy.loc[cs_and_math_dept, 'score'] * 0.10
df_copy.loc[cs_and_math_dept, 'score'] = df_copy.loc[cs_and_math_dept, 'score'] + increase_cs_and_math

score_col = df_copy['score']

increase_all_dept = score_col * 0.05
sum_increase_all_dept =  score_col + increase_all_dept
df_copy['score'] = np.where(cs_and_math_dept, score_col, sum_increase_all_dept)


df_copy['new_score'] = df_copy['score'].where(df_copy['score'] <= 100, 100)

df_copy['score_increase'] = df_copy['new_score'] - df_copy['previous_score']

new_df = df_copy[['student_id', 'department', 'previous_score', 'new_score', 'score_increase']]

print(new_df)
```

# Question 05: [ Marks 10 ]

Using Matplotlib, create a bar chart that shows the average score per department.
