# 2D line plot from pandas data

```python
df = pd.read_csv('enrollment_data.csv')
df
```

```python
#For single data

plt.xlabel('Years')
plt.ylabel('Count of enrollment')
plt.title('Student enrollment over the years')

plt.plot(df['Year'], df['Programming'])
```

```python
# For multiple data

plt.xlabel('Years')
plt.ylabel('Count of enrollment')
plt.title('Student enrollment over the years')

plt.plot(df['Year'], df['Programming'], label='Programming')
plt.plot(df['Year'], df['Digital Marketing'], label='Digital Marketing')
```

```python
# Adding labels using plt.legend

plt.xlabel('Years')
plt.ylabel('Count of enrollment')
plt.title('Student enrollment over the years')

plt.plot(df['Year'], df['Programming'], label='Programming')
plt.plot(df['Year'], df['Digital Marketing'], label='Digital Marketing')

plt.legend(loc='best')
plt.show()
```

```python
# Adding some visualization

plt.xlabel('Years')
plt.ylabel('Count of enrollment')
plt.title('Student enrollment over the years')

plt.plot(df['Year'], df['Programming'], label='Programming', color='red', linewidth=3, linestyle='dashed', marker='o')
plt.plot(df['Year'], df['Digital Marketing'], label='Digital Marketing', color='black', linewidth=3, linestyle='dotted', marker='o')

plt.legend()
plt.show()
```

```python
# Applying grid

# Adding some visualization

plt.xlabel('Years')
plt.ylabel('Count of enrollment')
plt.title('Student enrollment over the years')

plt.plot(df['Year'], df['Programming'], label='Programming', color='red', linewidth=3, linestyle='dashed', marker='o')
plt.plot(df['Year'], df['Digital Marketing'], label='Digital Marketing', color='black', linewidth=3, linestyle='dotted', marker='o')

plt.legend()
plt.grid()

plt.show()
```

# Scatter Plot

```python
students = pd.read_csv('student_IQdata.csv')
students
```

```python
# Relation between Study hours and IQ score

plt.xlabel('Study Hours')
plt.ylabel('IQ_Score')
plt.title('Relation between Study hours and IQ score')

plt.scatter( students['Study_Hour'], students['IQ_Score'])

plt.legend()
plt.grid()

plt.show()
```

```python
# Relation between shoe size and IQ score
plt.xlabel('Shoe Size')
plt.ylabel('IQ Score')
plt.title('Relation between shoe size and IQ score')

plt.scatter( students['Shoe_Size'], students['IQ_Score'])

plt.legend()
plt.grid()

plt.show()
```

```python
# Relation between chilling hour and IQ score
plt.xlabel('Chilling Hours')
plt.ylabel('IQ Score')
plt.title('Relation between shoe size and IQ score')

plt.scatter( students['Chilling_Hours'], students['IQ_Score'], color='red')

plt.legend()
plt.grid()

plt.show()
```

# Bar Chart

```python
data = {
    'Student': [ 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': [ 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [ 2, 5, 3, 7, 4, 8, 1],
    'Attendance': [ 60, 88, 70, 95, 80, 98, 55],
    'Grade': [ 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)

df
```

```python
# Gender wise

gender_group = df.groupby('Gender').size()

gender_group
```

```python
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Bar chart of genders among students')

plt.bar(gender_group.index, gender_group.values, color=['pink', 'black'], edgecolor='blue')
```

```python
# Based on real data

df = pd.read_csv('student_data.csv')

df
```

```python
# Status Wise

status = df.groupby('CompletionStatus').size()

plt.xlabel('Completion Status')
plt.ylabel('Count')
plt.title('Bar chart of course completion of students')

plt.bar(status.index, status.values, color=['green', 'orange', 'red'], edgecolor='black')
```

```python
# Location wise

status = df.groupby('Location').size()

plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Bar chart of location of students')

plt.bar(status.index, status.values, edgecolor='black')
```

# Pie Chart

```python
plt.pie(status, labels=status.index, autopct='%1.1f%%', explode=(0.1, 0.0, 0.0), shadow=True, startangle=90)

plt.show()
```
