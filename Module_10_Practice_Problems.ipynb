# Problem 06

```python
#creating a dataframe
datas = {
    "Name":["Ema", "John", 'Albert', 'James', "Nadia"],
    "Department":['Sales', 'Management', 'IT', 'HR', 'Clerk'],
    "Salary":[10000, 30000, 350000, 60000, 20000]
}

employees = pd.DataFrame(datas, index=[1,2,3,4,5])
employees
```

```python
# Ascending order by Salary
salary = employees.sort_values('Salary', ascending=True)
salary
```

```python
# Descending order by Name
Names = employees.sort_values('Name', ascending=False)
Names
```

```python
# Sort by Salary and Names
copy = employees.sort_values(['Department', 'Salary'], ascending=[True, False])
copy
```

# Problem 07

```python
#creating a dataframe
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Henry"],
    "Score": [85, 72, 90, 66, 39, 94, 33, 81]
}

chatro = pd.DataFrame(data, index=[1,2,3,4,5,6,7,8])
chatro
```

```python
# Score > 60
greater_than = chatro.loc[(chatro['Score'] > 60)]
greater_than
```

```python
# Score < 40.
less_than = chatro.loc[(chatro['Score'] < 40)]
less_than
```
