# Apply function on Df

```python
df
```

```python
# min max scaling

min_value = df['Total Marks'].min()
max_value = df['Total Marks'].max()

df['Scaled Marks'] = df['Total Marks'].apply(lambda x : (x - min_value) / (max_value - min_value))

df
```

```python
# Custom built functions

def grading_system(marks):
  if marks >= 260:
    return 'A+'
  elif marks >= 250:
    return 'A'
  else:
    return 'A-'

df['Grades'] = df['Total Marks'].apply(grading_system)

df
```

```python
def marking_system(df):
  a = df['Data Structure Marks'] * 2
  b = df['Python Marks'] * 3
  c = df['Algorithm Marks'] * 4

  return a+b+c

df['Exceptional Marks'] = df.apply(marking_system, axis=1)

df
```

```python
# Calculating total time taken to complete the course

df['Total Time Taken'] = df['FinishedDate'] - df['EnrollmentDate']

df
```

# Pandas Group By

```python
df = pd.read_csv('sample_data/student_completed_data.csv')
```

```python
# Instructor GroupBy

group = df.groupby('Instructor')

group
```

```python
# Drop columns

# df.drop(columns=['EnrollmentDate', 'FinishedDate'], inplace=True)
df.drop(columns=['StudentID', 'FullName', 'CompletionStatus'], inplace=True)
```

```python
# Instructor GroupBy

group = df.groupby('Instructor')

group.sum()
```

```python
group.min()
```

```python
group.max()
```

```python
# See first students

group.first()
```

```python
# See last students

group.last()
```

```python
group.describe()
```

```python
# Calculating average of instructor groupBy

avg_marks = group['Total Marks'].mean()

avg_marks
```

```python
# Filling the null values in col with average marks

df['Total Marks'] = df['Total Marks'].fillna(df['Instructor'].map(avg_marks))

df
```
