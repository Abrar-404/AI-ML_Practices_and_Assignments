# 2. A dataset has a mean of 50 and a standard deviation of 10. According to the Empirical Rule, find the range where approximately 95% of the data lies.

### For 95%
The formula is: Range = Mean ± (2 X Standard Deviation)

so,

Lower bound = 50 - (2 X 10) = 30

Upper Bound = 50 + (2 X 10) = 70

So the range is 30-70

# 3. A student scored 85 in an exam where the mean score is 70 and the standard deviation is 5. Calculate the Z-score of the student.

The formula of Z-score is:

Z = (x - u) / sd

So,

Z = (85-70)/5 = 3

# 4. Given Q1 = 25 and Q3 = 45, calculate the IQR and explain what it represents about the dataset.


Formula of IQR:

IQR = Q3 - Q1

so,

IQR = 45 - 25 = 20

So it represent that,

- It measures the spread of the middle 50% of data

- It is resistant to outliers (unlike range)
- It is used to detect outliers using the formula:

Lower fence = Q1 − 1.5×IQR = 25 − 30 = −5

Upper fence = Q3 + 1.5×IQR = 45 + 30 = 75

Any value outside −5 to 75 is an outlier

# 5. A dataset has a mean (μ) of 80 and a standard deviation (σ) of 5. Using the Empirical Rule, determine the range where approximately 68% of the data lies.

For 68%
The formula is:

Range = Mean ± (1 X Standard Deviation)

so,

Lower bound = 80 - (1 X 5) = 75

Upper Bound = 80 + (1 X 5) = 85

So the range is 75 - 85

# 6. The heights of students are normally distributed with a mean of 160 cm and a standard deviation of 8 cm. Find the range that contains approximately 95% of the students’ heights.

Range = Mean ± (2 X Standard Deviation)

so,

Lower bound = 160 - (2 X 8) = 144

Upper Bound = 160 + (2 X 8) = 176
